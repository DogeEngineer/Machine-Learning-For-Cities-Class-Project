```python
import pandas as pd
import numpy as np
```

```python
headers=['number_man','ID','OTAZ','DTAZ','TourID','age1','age2','age3','income1','income2','income3','income4','work','TT_auto','Cost_auto','Cost_Taxi','TT_Taxi','TT_Walk','TT_Bike','WHT_FAC3','TT_pt','access','egress','numTransfers','transferTime','pt_access','choice','TT_carpool','Cost_carpool','Cost_Bike','Cost_Walk','pt','Cost_pt','tsf','new','man','totalID','subID','Num_trips','delete1','delete2','delete3','delete4']
NonManhattanData=pd.read_csv('NonManhattanComplied-926_old.csv',header=None)
ManhattanData=pd.read_csv('ManhattanComplied-926_old.csv',header=None)
NonManhattanData.columns=headers
ManhattanData.columns=headers
NonManhattanData=NonManhattanData.drop(['delete1','delete2','delete3','delete4'],axis=1)
ManhattanData=ManhattanData.drop(['delete1','delete2','delete3','delete4'],axis=1)
```

```python
ManhattanData.columns
```

```python
FilterManhattanData=ManhattanData[['ID','age1','age2','age3','income1','income2','income3','income4','work','TT_auto','Cost_auto','Cost_Taxi','TT_Taxi','TT_Walk','TT_Bike','TT_pt','access','egress','numTransfers','transferTime','pt_access','TT_carpool','Cost_Bike','Cost_Walk','Cost_pt','WHT_FAC3','choice','OTAZ','DTAZ']].copy()
FilterManhattanData.head()
```

```python
FilterNonManhattanData=NonManhattanData[['ID','age1','age2','age3','income1','income2','income3','income4','work','TT_auto','Cost_auto','Cost_Taxi','TT_Taxi','TT_Walk','TT_Bike','TT_pt','access','egress','numTransfers','transferTime','pt_access','TT_carpool','Cost_Bike','Cost_Walk','Cost_pt','WHT_FAC3','choice','OTAZ','DTAZ']].copy()
FilterNonManhattanData.head()
```

```python
frames = [FilterManhattanData, FilterNonManhattanData]
result = pd.concat(frames)
del FilterNonManhattanData,FilterManhattanData,ManhattanData,NonManhattanData
```

```python
result=result.sort_values(by=['ID'])
result=result.reset_index()
```

```python
result=result.drop('index',axis=1)
result.head()
```

```python
del ManhattanData,NonManhattanData
```

```python
industry=pd.read_csv('IndustryCode.csv')
industry.head()
```

```python
industry=industry.drop(['index','Unnamed: 0'],axis=1)
```

```python
output=result.merge(industry, left_on='ID', right_on='Trip_Number')
output.head()
```

```python
output.columns
```

```python
eval10 = output.sample(frac=0.1)
```

```python
del output
```

```python
DrivingTimeMatrix=pd.read_csv('DrivingDistanceMatrix.csv',header=None)
DrivingCost=[]
for index, row in eval10.iterrows():
    DrivingCost.append(DrivingTimeMatrix[row['OTAZ']-1][row['DTAZ']-1]*0.58)
```

```python
eval10['Cost_driving']=DrivingCost
```

```python
eval10.to_csv('synthetic_population_eval.csv')
```
